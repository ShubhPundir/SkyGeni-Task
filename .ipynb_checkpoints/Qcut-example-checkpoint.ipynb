```python
import pandas as pd
import numpy as np

deal_amounts = [
    2500, 2700, 3200, 5000, 6200,
    7000, 12000, 13000, 15000, 18000,
    21000, 30000, 40000, 45000, 52000,
    60000, 70000, 80000, 90000, 100000
]

df = pd.DataFrame({"deal_amount": deal_amounts})
```

```python
df["log_acv"] = np.log1p(df["deal_amount"])
```

```python
df
```

```python
df["acv_bucket_original"] = pd.qcut(df["deal_amount"], q=4, labels=["Low", "Mid", "High", "Very High"])
df["acv_bucket_with_log"] = pd.qcut(df["log_acv"], q=4, labels=["Low", "Mid", "High", "Very High"])
```
