### Feature Engineering

```python
df["created_date"] = pd.to_datetime(df["created_date"])
df["closed_date"] = pd.to_datetime(df["closed_date"])

df["created_month"] = df["created_date"].dt.month
df["created_quarter"] = df["created_date"].dt.quarter
df["created_year"] = df["created_date"].dt.year
df['created_year_and_quarter'] = df["created_date"].dt.to_period("Q").astype(str)
df["created_weekday"] = df["created_date"].dt.weekday

df["closed_month"] = df["closed_date"].dt.month
df["closed_quarter"] = df["closed_date"].dt.quarter
df["closed_year"] = df["closed_date"].dt.year
df['closed_year_and_quarter'] = df["closed_date"].dt.to_period("Q").astype(str)
df["closed_weekday"] = df["closed_date"].dt.weekday

# Filter out the last quarter
latest_quarter = df['closed_year_and_quarter'].max()
df = df[df['closed_year_and_quarter'] != latest_quarter].copy()

# Time buckets (helps CRO thinking)
df["sales_cycle_bucket"] = pd.cut(
    df["sales_cycle_days"],
    bins=[0, 14, 30, 60, 90, np.inf],
    labels=["<2w", "2-4w", "1-2m", "2-3m", "3m+"]
)

# Binary outcome
df["is_won"] = (df["outcome"] == "Won").astype(int)

# ACV buckets (very important) --> Right skewed data, hence log transformed
df["acv_bucket"] = pd.qcut(
    df["deal_amount"], 
    # np.log1p(df["deal_amount"]), 
    q=4,
    labels=["Low", "Mid", "High", "Very High"]
)

# doesn't matter if I apply log or not in pd.qcut as log is a monotonic function (preserves order) 
# and qcut works by splitting thr ranked data into equal sized groups
# hence ranking order of values doesn't change post log applying
# NOTE: for strictly qcut into bucket applications, skip applying log as it is just additional compute, 
# look at examples/qcut-example.ipynb for sanity check
```

```python
# Basic sanity
print(True if df["deal_id"].is_unique else False)
print(True if set(df["outcome"].unique()) == {"Won", "Lost"} else False)
print(True if list((df["closed_date"] - df["created_date"]).dt.days) == list(df['sales_cycle_days']) else False)
```

### Distribution of ['deal_amount', 'sales_cycle_time']

```python
def histogram(data, bins:int = 20):
    plt.figure(figsize=(15, 6))

    counts, bin_edges, patches = plt.hist(
        data,
        bins=bins,
        edgecolor='black'
    )

    total = counts.sum()
    percentages = counts / total * 100

    # Bin centers for ticks and labels
    bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
    plt.xticks(bin_centers, rotation=45)

    # Add percentage labels above bars
    for count, pct, patch in zip(counts, percentages, patches):
        if count > 0:
            plt.text(
                patch.get_x() + patch.get_width() / 2,
                count,
                f'{pct:.1f}%',
                ha='center',
                va='bottom',
                fontsize=9
            )

    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.title('Distribution of Integer Values')

    plt.tight_layout()
    plt.show()
```

```python
win_rate_trend_detailed = (
    df.groupby(["closed_year","closed_quarter","closed_month"])
      .agg(
          deals=("deal_id", "count"),
          win_rate=("is_won", "mean"),
          successful_deals=("is_won", "sum"),
          revenue_generated=("deal_amount", "sum")
      )
      .reset_index()
)

win_rate_trend_detailed["win_rate"] *= 100

win_rate_trend_detailed
```

```python
win_rate_trend = (
    df.groupby(["closed_year_and_quarter"])
      .agg(
          deals=("deal_id", "count"),
          win_rate=("is_won", "mean"),
          successful_deals=("is_won", "sum"),
          total_revenue_generated=("deal_amount", "sum")
      )
      .reset_index()
)
win_rate_trend["win_rate"] *= 100

win_rate_trend
```

```python
# Calculate decrease over the last two quarters
last_two = win_rate_trend.tail(2)

decrease = last_two.iloc[0]["win_rate"] - last_two.iloc[1]["win_rate"]

print(f"Win rate has decreased by {decrease:.2f} percentage points over the last two quarters.")
```

```python
plt.figure(figsize=(15, 6))
plt.plot(
    win_rate_trend["closed_year_and_quarter"], 
    win_rate_trend["win_rate"], 
    marker="o"
)

# Add data labels
for x, y in zip(win_rate_trend["closed_year_and_quarter"], 
                win_rate_trend["win_rate"]):
    plt.text(x, y, f"{y:.1f}%", ha='center', va='bottom')

plt.title("Win Rate Trend Over Time")
plt.ylabel("Win Rate (%)")
plt.xlabel("Quarter")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Segment Level Leakage

```python
def win_rate_by_segment(df, col):
    return (
        df.groupby(["created_year_and_quarter",col])
          .agg(
              deals=("deal_id", "count"),
              win_rate=("is_won", "mean"),
              avg_acv=("deal_amount", "mean")
          )
          .assign(win_rate=lambda x: x["win_rate"] * 100)
          # .sort_values("win_rate")
    )

segments = ["region", "industry", "product_type", "lead_source"]

segment_tables = {
    seg: win_rate_by_segment(df, seg)
    for seg in segments
}
```
