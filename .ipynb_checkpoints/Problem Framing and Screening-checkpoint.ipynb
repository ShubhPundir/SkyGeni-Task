4. What assumptions are you making about the data or business?

2. Ai must answer segments that are responsible for dragging the win rate down, find pain points for failure to make the deal (who we sell to, how we sell to or where deals stall)

3. Metrics that matter most for diagnosing the win rate issues are
    - win rate
    - avg deal cycle time
    - ACV by segment

    - late-stage loss rate
    - deal aging index (actual cycle/ median cycle for that segment or group); differs for each group, example: sales_rep_id, product_type, lead_source, etc
4. Assumptions taken for the study
   - outcome labels are accurate
   - stage progression is consistent
   - sales process did not drastically change mid-period
   - no major pricing change occured
   - deal stages represent the last stage the deal reached before it was marked for won or lost

### Feature Engineering

```python
df["created_date"] = pd.to_datetime(df["created_date"])
df["closed_date"] = pd.to_datetime(df["closed_date"])

df["created_month"] = df["created_date"].dt.month
df["created_quarter"] = df["created_date"].dt.quarter
df["created_year"] = df["created_date"].dt.year
df['created_year_and_quarter'] = df["created_date"].dt.to_period("Q").astype(str)
df["created_weekday"] = df["created_date"].dt.weekday

df["closed_month"] = df["closed_date"].dt.month
df["closed_quarter"] = df["closed_date"].dt.quarter
df["closed_year"] = df["closed_date"].dt.year
df['closed_year_and_quarter'] = df["closed_date"].dt.to_period("Q").astype(str)
df["closed_weekday"] = df["closed_date"].dt.weekday

# Filter out the last quarter
latest_quarter = df['closed_year_and_quarter'].max()
df = df[df['closed_year_and_quarter'] != latest_quarter].copy()

# Time buckets (helps CRO thinking)
df["sales_cycle_bucket"] = pd.cut(
    df["sales_cycle_days"],
    bins=[0, 14, 30, 60, 90, np.inf],
    labels=["<2w", "2-4w", "1-2m", "2-3m", "3m+"]
)

# Binary outcome
df["is_won"] = (df["outcome"] == "Won").astype(int)

# ACV buckets (very important) --> Right skewed data, hence log transformed
df["acv_bucket"] = pd.qcut(
    df["deal_amount"], 
    # np.log1p(df["deal_amount"]), 
    q=4,
    labels=["Low", "Mid", "High", "Very High"]
)

# doesn't matter if I apply log or not in pd.qcut as log is a monotonic function (preserves order) 
# and qcut works by splitting thr ranked data into equal sized groups
# hence ranking order of values doesn't change post log applying
# NOTE: for strictly qcut into bucket applications, skip applying log as it is just additional compute, 
# look at examples/qcut-example.ipynb for sanity check
```

### Distribution of ['deal_amount', 'sales_cycle_time']

```python
def histogram(data, transform=None, bins:int = 20):
    plt.figure(figsize=(15, 6))

    if transform == 'log':
        counts, bin_edges, patches = plt.hist(
        np.log1p(df[data]),
        bins=bins,
        edgecolor='black'
    )
    else:
        counts, bin_edges, patches = plt.hist(
            df[data],
            bins=bins,
            edgecolor='black'
        )

    total = counts.sum()
    percentages = counts / total * 100

    # Bin centers for ticks and labels
    bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
    plt.xticks(bin_centers, rotation=45)

    # Add percentage labels above bars
    for count, pct, patch in zip(counts, percentages, patches):
        if count > 0:
            plt.text(
                patch.get_x() + patch.get_width() / 2,
                count,
                f'{pct:.1f}%',
                ha='center',
                va='bottom',
                fontsize=9
            )

    plt.xlabel('Value')
    plt.ylabel('Frequency')
    if not transform:
        plt.title(f'Distribution of {data}')
    else:
        plt.title(f'Distribution of {data} using {transform} transformation')
    plt.tight_layout()
    plt.show()
```

### Deal Stage

```python
df['deal_stage'].value_counts()
```

```python
df_grouped_deal_stage = (
    df.groupby("deal_stage")
      .agg(
          deals=("deal_id", "count"),
          win_rate=("is_won", "mean"),
          total_won=("is_won", "sum"),
          total_acv=("deal_amount", "sum"),
          avg_acv=("deal_amount", "mean"),
          avg_cycle=("sales_cycle_days", "mean"),
      )
      .reset_index()
)

df_grouped_deal_stage['total_lost'] = df_grouped_deal_stage['deals'] - df_grouped_deal_stage['total_won']
df_grouped_deal_stage
```

#### This dataset does NOT contain stage-by-stage movement history, rather it is the last recorded stage before the deal was closed, as each of them contains both lost and won outcomes

#### Since all deals are closed and the dataset only records a single deal_stage value, I interpret this field as the terminal stage reached before the outcome. Therefore, I analyze it as a point of exit variable rather than a funnel progression variable.

#### To properly diagnose funnel leakage, I would request stage transition history data (timestamped stage changes). Without it, we can only analyze terminal-stage loss concentration.

| Deal Stage      | Correct Interpretation (Final Exit Stage)                                                                                    |
| --------------- | ---------------------------------------------------------------------------------------------------------------------------- |
| **Qualified**   | The deal was disqualified or lost during early qualification and did not progress to formal solution evaluation.             |
| **Demo**        | The opportunity reached product demonstration but was lost during or immediately after the evaluation phase.                 |
| **Proposal**    | A formal proposal was delivered, but the opportunity was lost during pricing validation or internal stakeholder alignment.   |
| **Negotiation** | The deal advanced to commercial discussions (pricing, terms, procurement) but was ultimately lost before contract execution. |
| **Closed**      | The opportunity completed the full sales cycle and reached a formal closure event (either won or administratively closed).   |
